```python
import pandas as pd
import time
```

```python
start_time = time.time()
```

```python
cluster_df = pd.read_csv('preprocessed_topic4_cluster_matrix.csv')
```

```python
cluster_df.head()
```

```python
vp_df = pd.read_csv('topic4_verb_list_5000sents.csv')
```

```python
vp_df.head()
```

```python
for vp_idx in range(len(vp_df)):
  lemma_verb = vp_df.loc[vp_idx, 'lemma_verb']
  # print('vp: ', lemma_verb)

  for cluster_idx in range(len(cluster_df)):
    member_lemma_list = cluster_df.loc[cluster_idx, 'member_lemma'].replace('\n', '').strip('\'][\'').split('\' \'')

    if lemma_verb in member_lemma_list:
      vp_df.loc[vp_idx, 'cluster_title'] = cluster_df.loc[cluster_idx, 'cluster_title']

      '''
      print('cluster_idx: ', cluster_idx)
      print('cluster_title: ', cluster_df.loc[cluster_idx, 'cluster_title'])
      print('cluster list: ', member_lemma_list)
      '''
  # break
```
