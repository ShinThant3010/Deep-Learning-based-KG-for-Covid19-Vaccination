### 3. Merge result from full sentence (1) and ner data (2).

```python
import pandas as pd
import numpy as np
import json
import time

start_time = time.time()
```

```python
df_1 = pd.read_csv('corpus_without_tag_all.csv')
df_1.head()
```

```python
df_2 = pd.read_csv('corpus_with_tags_all.csv')
df_2.head()
```

```python
# df_2['word_id']=df_2['word_id'].astype(int)
```

```python
df = pd.merge(df_1, df_2, how = 'left', on=['doc_id', 'sent_id', 'word_id'])
```

```python
df.to_csv('corpus_for_ner.csv')
```

```python
end_time = time.time()

print('Estimate runtime: ', (end_time-start_time)/60, ' minutes.')
```
