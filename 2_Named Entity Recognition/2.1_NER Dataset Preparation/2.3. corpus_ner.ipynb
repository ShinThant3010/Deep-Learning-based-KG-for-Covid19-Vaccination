### 2.3. Add word_id to each entity

```python
import pandas as pd
import time

start_time = time.time()
```

```python
df = pd.read_csv('corpus_with_tags_explodeNeed_version.csv')
df.head()
```

```python
df['text_list'] = df['text_list'].apply(lambda x: x.strip('][').split(', '))
```

```python
df['flag'] = df['flag'].apply(lambda x: x.strip('][').split(', '))
```

```python
df['word_id'] = df['word_id'].apply(lambda x: x.strip('][').split(', '))
```

```python
df_explode = df.explode(['text_list', 'flag', 'word_id'])
```

```python
df_explode.to_csv('corpus_with_tags_all.csv')
```
