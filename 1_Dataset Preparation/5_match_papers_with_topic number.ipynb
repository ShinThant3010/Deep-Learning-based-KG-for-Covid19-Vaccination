### 5. Assign respective topic number to each document

```python
import pandas as pd
import time

start_time = time.time()
```

```python
topic_prob = pd.read_csv('document_topic_matrix.csv')
```

```python
topic_prob.head()
```

```python
topic_prob = topic_prob.iloc[: , 1:]
topic_prob.head()
```

```python
for i in range(6):
    name = 'Topic ' + str(i+1)
    topic_prob.rename(columns={topic_prob.columns[i]: name},inplace=True)
topic_prob.head()
```

```python
df = pd.read_csv('cord19_vaccine_2020_dataset_withBi-Trigram.csv')
```

```python
selected_topic = []
max_percent = []
```

```python
for idx in range(len(topic_prob)):
    for col in topic_prob.columns:
        if(topic_prob.loc[idx,col] == max(topic_prob.iloc[idx])):
            selected_topic.append(col)
            max_percent.append(max(topic_prob.iloc[idx]))
```

```python
topic_prob['Selected Topic'] = selected_topic
topic_prob['Max Percentage'] = max_percent
```

```python
topic_prob.head()
```

```python
df['Selected Topic'] = selected_topic
```

```python
df.to_csv('dataset_with_topic.csv')
```
